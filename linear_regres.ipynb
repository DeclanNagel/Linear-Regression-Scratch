```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# set seed to zero so it is reproducible
np.random.seed(0)

# np.random.seed(100,1) creates a 100x1 matrix, with values 0 - 1. i multiply it by 2 such that the range is 0 - 2, so there is more 
# variation in the data
X = 2 * np.random.rand(100,1)

# set 4 as the true bias and 3 as the true weight, with adding np.random.seed(100,1) to simulate the randomness encountered in the real world
y = 4 + 3 * X + np.random.randn(100,1)

y
```

```python
# scatter plot

plt.scatter(X, y, color="blue")
plt.xlabel("X (input feature)")
plt.ylabel("y (target output)")
plt.title("Randomly Generated Training Data")
plt.show()
```

```python
# initializing weight and bias randomly

w = np.random.randn()
b = np.random.randn()

print(f"weight: {w}")
print(f"bias: {b}")
```

```python
# set learning rate to a small number to ensure small steps
learning_rate = 0.01
# sets how many passed over the gradiest loop i will make
n_iter = 1000
# take the number of data points
n = X.shape[0]
```

```python
# gradient optimization

for i in range(n_iter):
    y_pred = (w * X) + b

    # partial derivitve for weights and biases 
    dw = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # weight and bias adjustment
    w -= learning_rate * dw
    b -= learning_rate * db

    # tracking updates
    if i % 100 == 0:
        # MSE
        loss = np.mean((y - y_pred)**2)
        print(f"Iteration {i}: Loss = {loss:.4f}")
```

```python
# Plot the original data
plt.scatter(X, y, color="blue", label="Data points")

# Plot the learned regression line
plt.plot(X, w * X + b, color="red", label="Learned Line")

plt.xlabel("X (input)")
plt.ylabel("y (output)")
plt.title("Linear Regression Fit After Training")
plt.legend()
plt.show()

print(f"Final learned weight (w): {w:.4f}")
print(f"Final learned bias (b): {b:.4f}")
```
